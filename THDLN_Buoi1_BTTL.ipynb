```python
# bai 11 : (3,2) va (2,3) khong cong duoc --> transpose 1 trong 2
ts1 = tf.random.uniform((3,2),minval=5,maxval=11,dtype=tf.int32)
ts2 = tf.random.uniform((2,3),minval=15,maxval=21,dtype=tf.int32)
print("ts1:\n",ts1.numpy())
print("ts2:\n",ts2.numpy())
# dao chieu tensor : tf.transpose
ts2_transpose = tf.transpose(ts2)
print("ts2 transpose:\n",ts2_transpose.numpy())
# chuyen phan tu o duong cheo thanh 0 de x+y van giu dg cheo x
# 1. xem phan tu duong cheo : tf.eye(n_dong ,n_cot,dtypee : bool )
diag_mask = tf.eye(num_rows=3,num_columns=2,dtype=tf.bool)
print("mask duong cheo:\n",diag_mask.numpy())
# 2. gan phan tu 0
final_ts2 = tf.where(diag_mask,0,ts2_transpose)
print("ts2 chuan hoa :\n",final_ts2.numpy())
# cong 2 tensor
result = tf.add(ts1,final_ts2)
print("ket qua:\n",result.numpy())
```
