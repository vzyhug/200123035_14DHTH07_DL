```python
# bai 11 : (3,2) va (2,3) khong cong duoc --> transpose 1 trong 2
ts1 = tf.random.uniform((3,2),minval=5,maxval=11,dtype=tf.int32)
ts2 = tf.random.uniform((2,3),minval=15,maxval=21,dtype=tf.int32)
print("ts1:\n",ts1.numpy())
print("ts2:\n",ts2.numpy())
# dao chieu tensor : tf.transpose
ts2_transpose = tf.transpose(ts2)
print("ts2 transpose:\n",ts2_transpose.numpy())
# chuyen phan tu o duong cheo thanh 0 de x+y van giu dg cheo x
# 1. xem phan tu duong cheo : tf.eye(n_dong ,n_cot,dtypee : bool )
diag_mask = tf.eye(num_rows=3,num_columns=2,dtype=tf.bool)
print("mask duong cheo:\n",diag_mask.numpy())
# 2. gan phan tu 0
final_ts2 = tf.where(diag_mask,0,ts2_transpose)
print("ts2 chuan hoa :\n",final_ts2.numpy())
# cong 2 tensor
result = tf.add(ts1,final_ts2)
print("ket qua:\n",result.numpy())
```

```python
# bai 12
#input
x_input = Input(shape=(3,3),dtype=tf.int32)
y_input = Input(shape=(3,3),dtype=tf.int32)
# output
x_output = Add()([x_input,y_input])
# model
model = Model(inputs=[x_input,y_input],outputs=x_output)
# data
n = int(input("nhap n: "))
x_data = tf.random.uniform((n,3,3),minval=1,maxval=10,dtype=tf.int32)
y_data = tf.random.uniform((n,3,3),minval=1,maxval=10,dtype=tf.int32)
print("x_data:\n",x_data.numpy())
print("y_data:\n",y_data.numpy())

# result
result = model([x_data,y_data])
print("result:\n",result.numpy())
```

```python
# bai 12
# input
x_input = Input(shape=(3,3),dtype=tf.int32)
y_input = Input(shape=(3,3),dtype=tf.int32)
# output
output_subtrack = Subtract()([x_input,y_input])
output_multiply = Multiply()([x_input,y_input])
# model
model = Model(inputs=[x_input,y_input],outputs=[output_subtrack,output_multiply])
# data
n = int(input("nhap n: "))
x_data = tf.random.uniform(shape=(1,3,3),minval=1,maxval=10,dtype=tf.int32)
y_data = tf.random.uniform(shape=(1,3,3),minval=1,maxval=10,dtype=tf.int32)
print("x_data:\n",x_data.numpy())
print("y_data:\n",y_data.numpy())
# result
result = model([x_data,y_data])
print("result subtrack:\n",result[0].numpy())
print("result multiply:\n",result[1].numpy())
```

```python
# bai 13 : model tinh chu vi HCN
# input
rect_lenght = Input(shape=(1,),dtype=tf.int32)
rect_width = Input(shape=(1,),dtype=tf.int32)
# output
output = (Add()([rect_lenght,rect_width]))*2
# model
model = Model(inputs=[rect_lenght,rect_width],outputs=output)
# data
input_lenght = int(input("nhap chieu dai: "))
input_width = int(input("nhap chieu rong: "))
rect_lenght_data = tf.constant([input_lenght],dtype=tf.int32)
rect_width_data = tf.constant([input_width],dtype=tf.int32)
# result
result = model.predict([rect_lenght_data,rect_width_data])
print("chu vi hinh chu nhat: ",result)
#
```
