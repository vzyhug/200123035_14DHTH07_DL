```python
import numpy as np
import pandas as pd
```

```python
# bai 1
# Tạo mảng ngẫu nhiên kích thước 3x3 trong khoảng [0, 20]
# Lưu ý: np.random.randint(low, high, size) thì 'high' sẽ không bao gồm trong kết quả
# nên ta dùng 21 để có thể xuất hiện số 20.
mang = np.random.randint(0, 21, size=(3, 3))

print("Mảng vừa tạo:")
print(mang)

# 2. Kiểm tra tất cả các phần tử có khác 0 hay không
# Biểu thức (mang != 0) sẽ trả về một mảng Boolean (True/False)
# Hàm np.all() trả về True nếu tất cả phần tử trong mảng Boolean đó là True
ket_qua = np.all(mang != 0)

print("\n--- Kết quả kiểm tra ---")
if ket_qua:
    print("Tất cả các phần tử trong mảng đều khác 0.")
else:
    print("Có ít nhất một phần tử bằng 0 trong mảng.")
```

```python
# bai2
# Tạo mảng ngẫu nhiên kích thước 3x3 trong khoảng [0, 20]
mang = np.random.randint(0, 21, size=(3, 3))

print("Mảng vừa tạo:")
print(mang)
# 2. Kiểm tra xem có TỒN TẠI phần tử nào khác 0 hay không
# Biểu thức (mang != 0) tạo mảng Boolean
# np.any() trả về True nếu có ít nhất một giá trị True
ton_tai_khac_0 = np.any(mang != 0)

print("\n--- Kết quả kiểm tra (any) ---")
if ton_tai_khac_0:
    print("Có tồn tại ít nhất một phần tử khác 0 trong mảng.")
else:
    print("Tất cả các phần tử đều bằng 0.")
```

```python
#bai 3
# Tạo dữ liệu mẫu
# Mảng 1 chiều
a1 = np.array([10, 20, 30])
b1 = np.array([15, 20, 25])

# Mảng 2 chiều (2x2)
a2 = np.array([[1, 5], [8, 12]])
b2 = np.array([[2, 4], [8, 10]])

def so_sanh_mang(arr1, arr2, label):
    print(f"--- So sánh mảng {label} ---")
    print(f"Mảng A:\n{arr1}")
    print(f"Mảng B:\n{arr2}")

    print("1. Greater (A > B):      ", np.greater(arr1, arr2))
    print("2. Greater Equal (A >= B):", np.greater_equal(arr1, arr2))
    print("3. Less (A < B):         ", np.less(arr1, arr2))
    print("4. Less Equal (A <= B):   ", np.less_equal(arr1, arr2))
    print("\n")

# Thực thi
so_sanh_mang(a1, b1, "1 chiều")
so_sanh_mang(a2, b2, "2 chiều")
```

```python
# bai 14
matrixA = np.random.randint(low=0,high=10,size=(4,3))
print("matrix A : \n",matrixA)
vtY = np.random.randint(low=0,high=10,size=3)
print("\nvector Y :",vtY)
result =[]
print("them vector Y vao tung dong cua ma tran A:")
# cach 1
print("cach 1:")
for i in range(0,matrixA.shape[0]):
    row = np.hstack((matrixA[i,:],vtY))
    result.append(row)
print(np.array(result))
# cach 2
print("cach 2:")
vtY_tiled = np.tile(vtY, (matrixA.shape[0], 1))
result = np.hstack((matrixA, vtY_tiled))
print(result)
#

#
```

```python
# bai 15
s1 = pd.Series([2, 4, 6, 8, 10])
s2 = pd.Series([1, 3, 5, 7, 10])

print("Series 1:\n", s1)
print("\nSeries 2:\n", s2)

# 1. Cong
res_add = s1 + s2
print("\nCong\n", res_add)

# 2. Tru
res_sub = s1 - s2
print("\nTru\n", res_sub)

# 3. Nhan
res_mul = s1 * s2
print("\nNhan\n", res_mul)

# 4. Chia
res_div = s1 / s2
print("\nChia\n", res_div)
```
