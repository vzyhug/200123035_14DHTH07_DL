```python
# bai 11 : (3,2) va (2,3) khong cong duoc --> transpose 1 trong 2
ts1 = tf.random.uniform((3,2),minval=5,maxval=11,dtype=tf.int32)
ts2 = tf.random.uniform((2,3),minval=15,maxval=21,dtype=tf.int32)
print("ts1:\n",ts1.numpy())
print("ts2:\n",ts2.numpy())
# dao chieu tensor : tf.transpose
ts2_transpose = tf.transpose(ts2)
print("ts2 transpose:\n",ts2_transpose.numpy())
# chuyen phan tu o duong cheo thanh 0 de x+y van giu dg cheo x
# 1. xem phan tu duong cheo : tf.eye(n_dong ,n_cot,dtypee : bool )
diag_mask = tf.eye(num_rows=3,num_columns=2,dtype=tf.bool)
print("mask duong cheo:\n",diag_mask.numpy())
# 2. gan phan tu 0
final_ts2 = tf.where(diag_mask,0,ts2_transpose)
print("ts2 chuan hoa :\n",final_ts2.numpy())
# cong 2 tensor
result = tf.add(ts1,final_ts2)
print("ket qua:\n",result.numpy())
```

```python
# bai 12:
# Hàm tạo tensor
def create_tensors():
    x = tf.random.uniform((3, 2, 2), minval=1, maxval=10, dtype=tf.float32)
    y = tf.random.uniform((3, 2, 2), minval=1, maxval=10, dtype=tf.float32)
    return x, y
#  2. Các hàm toán học
@tf.function
def add(x, y):
    return x + y

@tf.function
def subtract(x, y):
    return x - y

@tf.function
def multiply(x, y):
    return x * y

@tf.function
def divide(x, y):
    return x / y

#  3. Test
x, y = create_tensors()

print("x =\n", x.numpy())
print("y =\n", y.numpy())

print("\nCong:\n", add(x, y).numpy())
print("\nTru:\n", subtract(x, y).numpy())
print("\nNhan:\n", multiply(x, y).numpy())
print("\nChia:\n", divide(x, y).numpy())
```

```python
# bai 13 : model tinh tong 2 ma tran
#input
x_input = Input(shape=(3,3),dtype=tf.int32)
y_input = Input(shape=(3,3),dtype=tf.int32)
# output
x_output = Add()([x_input,y_input])
# model
model = Model(inputs=[x_input,y_input],outputs=x_output)
# data
n = int(input("nhap n: "))
x_data = tf.random.uniform((n,3,3),minval=1,maxval=10,dtype=tf.int32)
y_data = tf.random.uniform((n,3,3),minval=1,maxval=10,dtype=tf.int32)
print("x_data:\n",x_data.numpy())
print("y_data:\n",y_data.numpy())

# result
result = model([x_data,y_data])
print("result:\n",result.numpy())
```

```python
# bai 14 : model tinh hieu va tich cua ma tran
# input
x_input = Input(shape=(3,3),dtype=tf.int32)
y_input = Input(shape=(3,3),dtype=tf.int32)
# output
output_subtrack = Subtract()([x_input,y_input])
output_multiply = Multiply()([x_input,y_input])
# model
model = Model(inputs=[x_input,y_input],outputs=[output_subtrack,output_multiply])
# data
n = int(input("nhap n: "))
x_data = tf.random.uniform(shape=(1,3,3),minval=1,maxval=10,dtype=tf.int32)
y_data = tf.random.uniform(shape=(1,3,3),minval=1,maxval=10,dtype=tf.int32)
print("x_data:\n",x_data.numpy())
print("y_data:\n",y_data.numpy())
# result
result = model([x_data,y_data])
print("result subtrack:\n",result[0].numpy())
print("result multiply:\n",result[1].numpy())
```

```python
# bai 15 : model tinh chu vi HCN
# input
rect_lenght = Input(shape=(1,),dtype=tf.int32)
rect_width = Input(shape=(1,),dtype=tf.int32)
# output
output = (Add()([rect_lenght,rect_width]))*2
# model
model = Model(inputs=[rect_lenght,rect_width],outputs=output)
# data
input_lenght = int(input("nhap chieu dai: "))
input_width = int(input("nhap chieu rong: "))
rect_lenght_data = tf.constant([input_lenght],dtype=tf.int32)
rect_width_data = tf.constant([input_width],dtype=tf.int32)
# result
result = model.predict([rect_lenght_data,rect_width_data])
print("chu vi hinh chu nhat: ",result)
#
```

```python
# bai 16 : model tinh dien tich Hinh hoc Khong gian
# Input
a_input = Input(shape=(1,), dtype=tf.float32)
b_input = Input(shape=(1,), dtype=tf.float32)
c_input = Input(shape=(1,), dtype=tf.float32)

#  (a + b)
a_plus_b = Add()([a_input, b_input])

# Sxung quanh = (a + b) * c * 2
mul1 = Multiply()([a_plus_b, c_input])
Sxq = Multiply()([mul1, tf.constant([2.0])])

#  2ab
ab = Multiply()([a_input, b_input])
two_ab = Multiply()([ab, tf.constant([2.0])])

#  Stoan phan
Stp = Add()([two_ab, Sxq])

#Model
model = Model(inputs=[a_input, b_input, c_input], outputs=Stp)

#  Test
a = float(input("Nhap a: "))
b = float(input("Nhap b: "))
c = float(input("Nhap c: "))

a_data = tf.constant([[a]])
b_data = tf.constant([[b]])
c_data = tf.constant([[c]])

result = model.predict([a_data, b_data, c_data])

print("Dien tich toan phan:", result[0][0])
```

```python
from os import name
# Bai 18 :
# Co so ly thuyet : Hinh thang co 2 canh ben song song la Hinh binh hanh

# input
AD_input = Input(shape=(1,),dtype=tf.int32)
BC_input = Input(shape=(1,),dtype=tf.int32)
AB_input = Input(shape=(1,),dtype=tf.int32)
CD_input = Input(shape=(1,),dtype=tf.int32)

# AD=BC --> Subtract (AD,BC) = 0 , AB=CD --> Sutract(AB,CD)=0
diff_AD_BC = Subtract(name="Subtract_AD_BC")([AD_input,BC_input])
diff_AB_CD = Subtract(name="Subtract_AB_CD")([AB_input,CD_input])

# 0 --> Sai , 1-->Dung
check_zero = Lambda(lambda x: tf.cast(tf.math.equal(x,0),tf.int32),name="Is_zero")
cond_AD_BC = check_zero(diff_AD_BC)
cond_AB_CD = check_zero(diff_AB_CD)
# output : 0*n=0 , 1*1=1
output = Multiply(name="Output")([cond_AD_BC,cond_AB_CD])
# model
model = Model(inputs=[AD_input,BC_input,AB_input,CD_input],outputs=output)
#model.summary()
# data
AD = int(input("nhap AD: "))
BC = int(input("nhap BC: "))
AB = int(input("nhap AB: "))
CD = int(input("nhap CD: "))
AD_data = tf.constant([[AD]],dtype=tf.int32)
BC_data = tf.constant([[BC]],dtype=tf.int32)
AB_data = tf.constant([[AB]],dtype=tf.int32)
CD_data = tf.constant([[CD]],dtype=tf.int32)
# result
result = model.predict([AD_data,BC_data,AB_data,CD_data])
gia_tri_model = result[0][0]

print(f"--- THÔNG SỐ: AD={AD_data}, BC={BC_data}, AB={AB_data}, CD={CD_data} ---")

if gia_tri_model == 1:
    # Model trả về 1 đồng nghĩa với việc cả 2 điều kiện đều thỏa mãn
    print("Trường hợp 1: AD = BC và AB = CD.")
    print("=> Kết luận: AD // BC (Đây là Hình bình hành).")

else:
    # Model trả về 0, nghĩa là ít nhất một trong hai cặp cạnh KHÔNG bằng nhau.
    # Ta dùng if/elif/else để rẽ nhánh tìm nguyên nhân cụ thể:

    if AD_data == BC_data and AB_data != CD_data:
        print("Trường hợp 2: Cạnh bên AD = BC, nhưng đáy AB khác CD.")
        print("=> Kết luận: AD KHÔNG song song với BC (Đây là Hình thang cân).")

    elif AD_data != BC_data and AB_data == CD_data:
        print("Trường hợp 3: Đáy AB = CD, nhưng cạnh bên AD khác BC.")
        print("=> Kết luận: Dữ liệu bị lỗi logic!")
        print("   (Lý thuyết: Hình thang có 2 đáy bằng nhau thì buộc phải là hình bình hành, nên AD phải bằng BC).")

    else:
        # Trường hợp còn lại: ad != bc và ab != cd
        print("Trường hợp 4: Cạnh bên AD khác BC và đáy AB khác CD.")
        print("=> Kết luận: AD KHÔNG song song với BC (Đây là Hình thang thường).")
#
```
